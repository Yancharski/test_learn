```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных из CSV-файла в DataFrame
data = {'retention_time': [7.6, 9.6, 11.6, 13.6, 15.6, 17.6, 19.6, 21.6, 23.6, 25.6, 27.6, 29.6, 31.6],
        'intensity': [0, 0.111, 0.124, 0.012, 0.004, 1.145, 0.386, 0.066, 0.024, 0.018, 0.064, 0.055, 0.043]}

# Создание DataFrame
df = pd.DataFrame(data)

# Построение графика
plt.plot(df['retention_time'], df['intensity'])
plt.xlabel('Объём')
plt.ylabel('Оптическая плотность')
plt.title('График обессоливания белка')
plt.show()
```
